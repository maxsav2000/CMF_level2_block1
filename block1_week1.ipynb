```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('tb.xlsx')
df
```

```python
def normalize_data(df):
    ret = df
    for i in range(len(df)):
        q = df[i].split('/')
        month = q[0]
        day = q[1]
        if len (month) == 1:
            month = '0' + str(month)
        if len (day) == 1:
            day = '0' + str(day)
        new_data = str(20) + str(q[2]) + '-' + str(month) + '-' + str(day)
        ret[i] = new_data
    return ret
```

```python
df['StartDate'] = normalize_data(df['StartDate'])
df['EndDate'] = normalize_data(df['EndDate'])
```
